## **1.1** Investigación y preámbulo sobre aprendizaje supervisado.
- **Aprendizaje supervisado:**
El aprendizaje supervisado es el tipo más común de algoritmos de aprendizaje automático, se utiliza para clasificar datos no vistos en categorías establecidas y pronosticar tendencias y cambios futuros como un modelo predictivo. En pocas palabras, hay un modelo generado que aprende la relación entre los datos de entrada y salida etiquetados para predecir, o sea intenta dar valores de respuesta posibles para un nuevo conjunto de datos. Lo anterior se realiza mediante la utilización de un training dataset o set de datos conocido, para entrenar un algoritmo con un cierto input data denominado features, este con resultados conocidos(ouput data).

- **Problema de clasificación:**
Este se trata de situaciones donde se busca clasificar dentro de un problema de modelado predictivo, es decir, donde se predice una etiqueta para un ejemplo dado de datos de entrada. Por ejemplo, un problema de clasificación es cuando la variable de salida es una categoría, como "rojo" o "azul" o "enfermedad" y "sin enfermedad". Un modelo de clasificación intenta sacar alguna conclusión de los valores observados, básicamente que a partir de cierto input data se puede saber a qué categoría pertenece. Otros ejemplos sería si un mensaje es Spam o no, etc. Hay distintos tipos de clasificación de estos problemas o casos, entre ellos está; clasificación binaria (tengo cancer o no), clasificación multiclase (reconocimiento facial, es decir tu cara frente a otras), clasificación multietiqueta(predicción acerca del genero y contenido de una película).

## **1.2** Descripción breve de los algunos clasificadores
- **Árbol de decisión:**
El árbol de decisiones es una técnica de aprendizaje supervisado. Se trata de un clasificador con estructura de árbol, donde los nodos internos representan las características de un conjunto de datos, las ramas representan las reglas de decisión y cada hoja representa el resultado. Básicamente, la idea general de este clasificador es imitar la forma en que humanos tomamos decisiones, es decir se usan las features del data set para formular preguntas del tipo 'Sí' o 'No' y así dividir continuamente el conjunto de datos hasta aislar todos los puntos de datos que pertenecen a cada clase, siendo así clasificadas.

- **Suport vector Machine (SVM):**
Por su nombre en español, máquinas de vectores de soporte, este también es un algoritmo de aprendizaje automático supervisado, aquí tenemos que trazar cada elemento de datos como un punto en un espacio n-dimensional (donde n sería la cantidad de features que tiene el dataset) y el valor de cada feature es el valor de una coordenada particular. Luego, la clasificación se haría encontrando el hiperplano(vector) que diferencia ambas dos clases que se buscan estudiar. Entonces sería como la frontera que segrega mejor a dos clases.

- **Naive Bayes:**
Los clasificadores Naive Bayes son una colección de algoritmos de clasificación basados en el Teorema de Bayes.
A partir del teorema de Bayes, podemos encontrar la probabilidad de que ocurra A, dado que ha ocurrido B. Aquí, B es la evidencia y A es la hipótesis. Para lo anterior se supone que los predictores/features son independientes, es decir, la presencia de una feature en particular no afecta a la otra. Por eso se llama 'naive' o 'ingenuo' en español, y es esta la mayor desventaja de este método, ya que en la vida real existen dependencias.
Sobre su funcionamiento, este clasificador devuelve la probabilidad de que un punto de prueba pertenezca a una clase en lugar de la etiqueta del punto de prueba.

- **Vecinos mas ceranos (KNN):**
Nuevamente hablamos de un algoritmo de aprendizaje automático supervisado que almacena los casos disponibles y clasifica nuevos casos en función de una medida de similitud (por ejemplo, funciones de distancia; euclideana, manhattan, etc). Este algoritmo presenta desventaja cuando el volumen de datos es mayor, ya que de eso depende el reconocimiento y se hace lento mientras más datos haya que analizar.
El funcionamiento de este clasificador sería tal que si trazamos puntos de un dataset en un gráfico, podemos ubicar algunos grupos o clusters. Ahora, dado un punto sin clasificar, podemos asignarlo a un grupo observando a qué grupo pertenecen sus vecinos más cercanos, de ahí viene el nombre de este clasificador.


## **1.3** Definición sobre las métricas de desempeño de interés
- **Accuracy:**
Una métrica de desempeño que habla de la exactitud, o sea, el porcentaje total de elementos clasificados correctamente.

- **Falso positivo, falso negativo, verdadero positivo, verdadero negativo:**
Dentro del contexto de predicciones y análisis, un falso positivo sería cuando cuando la clase real del punto de datos era 0 (Falso) y el pronosticado es 1 (Verdadero), un ejemplo sería tener una enfermedad cuando en realidad no se tiene. Por otro lado, el falso negativo es que la clase real del punto de datos era 1 (Verdadero) y el valor predicho es 0 (Falso). 
Un verdadero positivo sería cuando la clase real del punto de datos era 1 (Verdadero) y la predicha también es 1 (Verdadero). Finalmente un verdadero negativo sería que la clase real del punto de datos fue 0 (Falso) y el pronosticado también es 0 (Falso).

- **Matriz de confusión:**
Se trata de una tabla que describe el rendimiento de un modelo supervisado de Machine Learning o Aprendizaje de Máquina en el dataset, donde se desconocen los verdaderos valores, a su vez, el nombre viene de que con ella es fácil detectar donde el sistema confunde 2 clases.

- **Precisión y Recall:**
El recall sería básicamente una métrica que ayuda con el número de elementos identificados correctamente como positivos del total de positivos verdaderos, es decir el número de verdaderos positivos sobre la suma de verdaderos positivos más los falsos negativos.
Por otro lado, la precisión estaría ligada al número de elementos identificados correctamente como positivo de un total de elementos identificados como positivos, es decir, los verdaderos positivos sobre la suma de verdaderos positivos y la de falsos positivos.
Ambas métricas nos servirían para la idea de minimizar o falsos negativos o positivos, así sirviéndonos para ampliar la idea de precisión del análisis.

## **Parte 3**

**Parte 3.1** Se obtienen subconjuntos de para realizar entrenamiento y prueba (en esta tarea no se utilizará conjunto de validación). La distribución es la siguiente: 30% para prueba y 70% para entrenamiento. 
**Nota**: Se utilizará train_test_split

```python
# Se carga librería de sklearn
from sklearn.model_selection import train_test_split

# Se obtienen los valores del dataframe
dato = datario.values

# Se separan los datos en ouput e input según la variable de estudio vista anteriormente.
x, y = dato[:, :-1], dato[:, -1]
print(x.shape, y.shape)

# Se separan entre valores de prueba y de entrenamiento, esto utilizando train_test_split
# 01:
#x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, shuffle=False) # Shuffle falso es que no sea randomizado
#print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)

# 02:
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=1)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape) #Hago print a lo obtenido
```

**Parte 3.2** Entrenamiento de diferentes arboles de decisión. 

Se prueba con diferentes parámetros:
   - Criterio de split
   - Máxima profundidad

```python
# Importamos el métricas y clasificador:
from sklearn import metrics #Métrica como accuracy para medir
from sklearn.tree import DecisionTreeClassifier #Arbol de decisión
clasificadorarbol = DecisionTreeClassifier(criterion="gini", random_state=0)  
clasificadorarbol.fit(x_train, y_train)

# Segundo caso cambiando el criterio y la profundidad:
clasificadorarbol2 = DecisionTreeClassifier(criterion="entropy", max_depth=10)
clasificadorarbol2.fit(x_train, y_train)
prediarbol = clasificadorarbol2.predict(x_test)

# Tercer caso solo con cambio en max_depth
clasificadorarbol3 = DecisionTreeClassifier(max_depth=10)
clasificadorarbol3.fit(x_train, y_train)

# Luego en la predicción, tenemos que la precisión es:
def accuracy(caso):
 u_predientrenado = caso.predict(x_train)
 print("Precisión en el caso de entrenamiento:", metrics.accuracy_score(y_train, u_predientrenado))
 v_prediprueba = caso.predict(x_test) 
 print("Precisión en el caso de prueba:", metrics.accuracy_score(y_test, v_prediprueba))

print('Casos:')
accuracy(clasificadorarbol)
print('casos con valores distintos:')
# Caso con mayor cambio en el valor de max depth y otro criterio:
accuracy(clasificadorarbol2)
# Caso solo con cambio en  profundidad
accuracy(clasificadorarbol3)
```

Podemos notar que este clasificador a entrenado y obtenido valores del 100%, mientras que los valores de prueba no lo son, y dentro de esto mismo, para los valores casi por default se obtiene un 68% y para los valores con cambios en la profundidad tenemos una mejora en la precisión. Max_depth s básicamente el número de nodos desde la raíz hasta la hoja más lejana, variar este valor cambia el tamaño del arbol y por ende su consideración de datos, asimismo evitar outliers o un análisis demasiado disimil.

**Parte 3.3** Entrenamiento de un clasificador SVM

**Nota**: from sklearn import svm

```python
# Se importa el clasificador
from sklearn import svm
clasificadorsvm = svm.SVC(kernel='linear')
clasificadorsvm.fit(x_train, y_train)
predisvm = clasificadorsvm.predict(x_test)

# Caso polinomial
clasificadorsvm2 = svm.SVC(kernel='poly', degree=2)
clasificadorsvm2.fit(x_train, y_train)
predisvm2 = clasificadorsvm2.predict(x_test)

print('Precisión casos:')
accuracy(clasificadorsvm)
print('caso con grado 2;')
accuracy(clasificadorsvm2)
```

Podemos notar que a diferencia del clasificador anterior, la lectura y precisión en los datos de entrenamiento no es perfecta, sino que de 0.86 y 0.89 respectivamente para el caso linear y de grado 2. La precisión del caso polinomial de grado 2 parece ser mayor a la del caso linear.


Como se había vito con anterioridad, el clasificador SVM busca separar los datos a través de una línea, esta tiene que ser la más lejana respecto a los puntos más cercanos hacia los valores de la otra clase. Los Kernel de mayor grado producen un límite de decisión/línea más flexible.

**Parte 3.4** Entrenamiento de un clasificador Naive Bayes

**Parte 3.5** Entrene un Clasificador KNN
 - Distancia
 - numero de vecinos (n_neiborgh)

```python
# Se carga la librería
from sklearn.neighbors import KNeighborsClassifier
# Se genera el entrenamiento y la lectura:
clasificadorvecino = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)  #Caso default
clasificadorvecino.fit(x_train, y_train)

# Caso con variaciones en el número de vecinos:
clasificadorvecino2 = KNeighborsClassifier(n_neighbors=10, metric='minkowski', p=2)
clasificadorvecino2.fit(x_train, y_train)

# Caso con variaciones en el tipo de distancia:
clasificadorvecino3 = KNeighborsClassifier(n_neighbors=5, metric='manhattan')
clasificadorvecino3.fit(x_train, y_train)
predivecino = clasificadorvecino3.predict(x_test)


# Vemos la precisión:
print('Casos:')
accuracy(clasificadorvecino)
print('caso con 10 vecinos:')
accuracy(clasificadorvecino2)
print('caso con metrica manhattan:')
accuracy(clasificadorvecino3)
```

Se puede notar variaciones en la precisión con el cambio de las variables a estudiar, el número mayor de vecinos disminuyó la precisión del proceso, esto podría deberse a la aparición de más elementos disrruptores en el análisis. Asimismo, la métrica minkowski se vio superada por la manhattan, esto podría deberse a ser un camino más acorde a la distribución de los datos, lo que nos dio el aumento de precisión.

```python
#Haremos una función para poder ver cada una de las metricas pedidas.
def metricando(predi): 
 # Hacer print de las métricas precision, recall y matriz de confusión:
 # Se ocupará el modulo para precisión y recall, mientras que la de confusión se hará aparte: 
 precision, recall, fscore, support = metrica(y_test, predi)

 print('precision: {}'.format(precision))
 print('recall: {}'.format(recall))

 # Luego, la matriz sería:
 matriz = confusion_matrix(y_test, predi, labels=["R", "M"])

 # Se ocupará seaborn para hacer plot tal de tener una mejor visualización
 fx = seab.heatmap(matriz, annot=True)
 fx.set_title('Matriz de confusión'); #Aquí inician los labels
 fx.set_xlabel('Predicciones')
 fx.set_ylabel('Valores factuales ');

 # Se hace label de las cajas
 fx.xaxis.set_ticklabels(['Falso','Verdadero'])
 fx.yaxis.set_ticklabels(['Falso','Verdadero'])

 atlas.show() #Se da show con atlas de plt.

print('Caso Arbol De Decisiones:')
metricando(prediarbol)
print('Caso SVM:')
metricando(predisvm2)
print('Caso Naive Bayes:')
metricando(predibayes)
print('Caso Vecino Más Cercano:')
metricando(predivecino)
```

**Parte 4.2** Ahora, se realizará un gráfico comparativo entre los clasificadores analizados y se discutirá las diferentes matrices de confusión.

**Parte 4.3** Validaremos los modelos obtenidos

```python
# Para validar haremos el modelado de nuevo y graficaremos, los valores de las métricas nos dirán qué clasificador es superior.
# Lista para automatizar el tema del gráfico y la reevaluación:
clasificadormodelo = {}

# Arbol de decisión
from sklearn.tree import DecisionTreeClassifier
clasificadormodelo['Arbol de Decision'] = DecisionTreeClassifier(criterion="entropy", max_depth=10) #Se utiliza la de mayor accuracy, es decir max_depth=10

# Estudio SVM
from sklearn import svm
clasificadormodelo['Support Vector Machines'] = svm.SVC(kernel='poly', degree=2) #Se escoge poly con grado 2

# Estudio Naive Bayes
from sklearn.naive_bayes import GaussianNB
clasificadormodelo['Naive Bayes'] = GaussianNB(var_smoothing=0.1) #Gaussiana con suavizado

# Estudio KNN
from sklearn.neighbors import KNeighborsClassifier
clasificadormodelo['K-Vecinos Más Cercanos'] = KNeighborsClassifier(n_neighbors=5, metric='manhattan') #Manhattan es la del caso de mayor accuracy

# Se importa modulo para mayor simplicidad
from sklearn.metrics import accuracy_score, precision_score, recall_score

#Lista de los valores:
accuracy, precision, recall = {}, {}, {}

#Para cada input necesario para cada clasificador:
for acción in clasificadormodelo.keys():
    
    # Se hace el fit:
    clasificadormodelo[acción].fit(x_train, y_train)
    
    # Prediccion:
    prediccion = clasificadormodelo[acción].predict(x_test)
    
    # Calcular Accuracy, Precision y Recall:
    accuracy[acción] = accuracy_score(prediccion, y_test)
    precision[acción] = precision_score(prediccion, y_test, pos_label='M')
    recall[acción] = recall_score(prediccion, y_test, pos_label='M')

#Se genera un dataframe de las métricas, esto para compararlas con un plot.
compara = pd.DataFrame(index=clasificadormodelo.keys(), columns=['Accuracy', 'Precision', 'Recall']) #Columnas y valores
compara['Accuracy'] = accuracy.values()
compara['Precision'] = precision.values()
compara['Recall'] = recall.values()

compara

#Se hace el plot:
grafi  = compara.plot.bar(rot=45)
grafi.legend(ncol= len(clasificadormodelo.keys()), bbox_to_anchor=(0, 1), loc='lower left', prop={'size': 14})
plt.tight_layout()
```
